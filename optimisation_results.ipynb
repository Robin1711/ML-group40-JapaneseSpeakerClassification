```python
import matplotlib.pyplot as plt
import skopt
import pickle
from skopt import plots
from glob import glob

folders=["RF"]#"BC", "DTC", "ESN", "KNN", "RF", 
for folder in folders:
    result_files = glob("opt/"+folder+"/*")
    res = []
    for result_file in result_files:
        print(result_file)
        results = skopt.load(result_file)
        print('Best Validation Accuracy: {}'.format(-1.0 * results.fun))
        print('Best Params: {}'.format(results.x))     
        res.append((result_file, results))
    fig, ax = plt.subplots(dpi=300)
    plots.plot_convergence(*res, ax.legend(loc=7, numpoints=1))
```
